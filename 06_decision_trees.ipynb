# 6.1 결정 트리 학습과 시각화

```
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:] # 꽃잎 길이와 너비
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth = 2, random_state = 42)
tree_clf.fit(X, y)
```

```
import os
from sklearn.tree import export_graphviz

export_graphviz(
    tree_clf, 
    out_file = os.path.join("iris_tree.dot"),  # image-path -> os.path.join
    feature_names = iris.feature_names[2:], 
    class_names = iris.target_names, 
    rounded = True,
    filled  = True
)
```

```
# 책에 없지만 이미지 출력을 위한 코드(그림 6-1, p231)
from graphviz import Source
Source.from_file(os.path.join("iris_tree.dot"))
```

# 6.3 클래스 확률 추정

```
tree_clf.predict_proba([[5, 1.5]])
```

# 6.8 회귀

```
import numpy as np

np.random.seed(42)

m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m, 1) / 10
```

```
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth = 2, random_state = 42)
tree_reg.fit(X, y)
```

```
# 책에 없지만, 이미지 출력을 위해 코드 추가
export_graphviz(
    tree_reg,
    out_file = os.path.join("regression_tree.dot"),
    feature_names = ["x1"], 
    rounded = True,
    filled  = True
)

Source.from_file(os.path.join("regression_tree.dot"))
```
