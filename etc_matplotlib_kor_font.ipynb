# 한글 깨질 때

```
import matplotlib.pyplot as plt

plt.figure(figsize = (5, 5))
plt.plot([0, 1], [0, 1], label = '한글 폰트')
plt.legend()
plt.show()
```

# 런타임 다시시작 및 matplotlib 폰트 변경

```
import matplotlib.pyplot as plt
plt.rc('font', family = 'NanumBarunGothic')

plt.figure(figsize = (5, 5))
plt.plot([0, 1], [0, 1], label = '한글 폰트')
plt.legend()
plt.show()
```
